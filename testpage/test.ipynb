```python
for result in results:
    #탐지된 객체이름
    className = result.boxes.cls.tolist()
    #탐지된 객체 정확도
    acc = result.boxes.conf.tolist()
    #탐지된 객체 xy좌표
    xy_list = result.masks.xyn



        
```

```python
print(className)
print(acc)
```

```python
print(xy_list)
```

```python
for index, (coords, class_val, acc_val) in enumerate(zip(xy_list, className, acc)):
    x_coords = [coord[0] for coord in coords]
    y_coords = [coord[1] for coord in coords]
    
    print(f"Object {index} - Class: {class_val}, Accuracy: {acc_val}, x coordinates: {x_coords} ,y coordinates: {y_coords}")
```
